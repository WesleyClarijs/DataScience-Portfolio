```python
import seaborn as sns
import pandas as pd
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins.groupby(by='island').mean()
```

```python
sns.barplot(x='island', y='bill_length_mm', data=penguins)
```

In bovenstaande barplot is te zien dat er een verschil zit tussen de 'bill length' en op welk eiland de desbetreffende pinguin leeft. Uit bovenstaande grafiek is af te lezen dat de pinguins die op het eiland Torgersen leven de kleinste 'bill length' hebben.

```python
sns.barplot(x='island', y='bill_depth_mm', data=penguins)
```

In bovenstaande barplot is te zien dat de pinguins die op Biscoe island leven, de kleinste 'bill depth' hebben. In tegenstelling tot de eerste grafiek, is er nu een duidelijk verschil tussen Biscoe island en de andere eilanden waar de pinguins leven.

```python
sns.barplot(x='island', y='flipper_length_mm', data=penguins)
```

In bovenstaande barplot is een duidelijk verschil te zien tussen de pinguins op Biscoe Island en de Dream en Torgersen eilanden. 

```python
sns.barplot(x='island', y='body_mass_g', data=penguins)
```

Als laatste is in bovenstaand barplot duidelijk dat de zwaarste pinguins op Biscoe island leven en de pinguins op de eilanden Torgersen en Dream een aanzienlijk lagere 'body mass' hebben. Bij elke barplot is een verschil te zien tussen de pinguins op de verschillende eilanden. De verschillen zijn niet altijd gelinkt aan hetzelfde eiland waardoor er per eiland verschillende conclusies getrokken kunnen worden.
