```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()
```

```python
sns.pairplot(penguins)
```

Uit bovenstaande pairplot kan ik persoonlijk opmaken dat er 2 tot 3 clusters aanwezig zijn. Waarschijnlijk zijn dit de verschillen tussen de species van de penguins. De verschillen tussen de species zijn op bepaalde punten zeer duidelijk zichtbaar en bij andere gegevens zijn de species zo goed als gelijk en delen zij dezelfde eigenschappen. Verder in dit notebook zal er gekeken worden of deze verwachting dat deze clusters de verschillende species bevatten, daadwerkelijk klopt.

```python
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
km = KMeans(n_clusters = 3, random_state=43).fit(penguins[features])
```

```python
penguins['cluster'] = km.predict(penguins[features])
```

```python
penguins.head()
```

```python
penguins.cluster.value_counts()
```

```python
sns.pairplot(penguins, hue='cluster')
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
metrics.silhouette_score(penguins[features], km.labels_, metric='euclidean')
```

```python
sns.pairplot(penguins, hue="species")
plt.show()
```

```python
contingency_table = penguins.groupby(['species','cluster']).size().unstack('cluster', fill_value=0)
contingency_table
```

Zoals hierboven te zien heeft elke cluster een duidelijke specie welke bovenaan staat. Echter zijn er ook gemengde resultaten, wat wil zeggen dat niet elke cluster duidelijk n specie bevat. Dit wil zeggen dat het grootste deel van de cluster daadwerkelijk n specie bevat, maar dat enkele andere penguins van een andere specie dezelfde eigenschappen, of in de buurt komt van de eigenschappen, van de desbetreffende overheersende specie. Dit zijn de uischieters die buiten het gemiddelde vallen van een bepaalde specie. Kortweg, het grootste deel van de species heeft duidelijke kenmerken die passen bij alleen die desbetreffende specie, er zijn uitschieters van andere species, dit wil zeggen dat hun eigenschappen niet kenmerkend zijn voor hun soort waardoor er verschillende species binnen 1 cluster aanwezig kunnen zijn.
